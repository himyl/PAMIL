```python
import pandas as pd
dataset = ["citeseer", "pubmed"]
for name in dataset:
    with open("./log_M1_epoch200/best_metric_{}_M1_epoch200.txt".format(name), "r") as f:
        data = f.read()
    data = data.split("\n")
    data = [_.strip() for _ in data if len(_) > 0]
    data = [_.split(",") for _ in data]
    data = [{i.split("=")[0].strip(): float(i.split("=")[1].strip()) for i in line} for line in data]
    data = pd.DataFrame(data)
    data.drop_duplicates(inplace=True)
    data.sort_values(by="Acc", ascending=False, inplace=True)
    data.to_csv("metric_{}.csv".format(name), index=0)
    print(name)
    # print(data["Acc"].std(), data["NMI"].std(), data["ARI"].std())
```
